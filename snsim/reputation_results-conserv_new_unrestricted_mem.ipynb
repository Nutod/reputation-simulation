```python
import pandas as pd
from copy import deepcopy 

allcols = ['code','folder','log','fullnorm','conserv','default','needs','downrating','decayed','period',
           'precision','recall','f1','pearson_by_good', 'pearsong_by_good','pearsonb_by_good',
           'loss_to_scam','profit_from_scam','inequity','utility', 'market_volume']
#'precision','recall','f1',

columns = ['log','default','conserv','needs','downrating','decayed','fullnorm','period','folder','code']


alist = [
['false','0.5','0.5','equil','false','0.5','true','1', 'conserv_new_unrestricted_mem']
]
 
#codelist = ['r_20_1']


codelist = ['r_20_1','r_10_1','r_5_1','r_1_1','r_20_0.5','r_10_0.5',
            'r_5_0.5','r_1_0.5','r_20_0.1','r_10_0.1','r_5_0.1','r_1_0.1']


testfiles = [
             "discrete_rank_tests.tsv",
             "correlation_by_good_tests.tsv",
             "scam_loss_tests.tsv",
             "utility_tests.tsv",
             "inequity_tests.tsv", 
             "market_volume_tests.tsv",
             "price_variance_tests.tsv",
             "correlation_tests.tsv",
             "continuous_rsmd_tests.tsv",
             "continuous_rsmd_by_good_tests.tsv",
             "scam_profit_tests.tsv",
            ]
    
```

```python
dflist = []
for row in alist:
    print(row)
    runslist = []
    for code in codelist:
        copy = deepcopy(row)
        copy.append(code)
        runslist.append(copy)
        runs = pd.DataFrame(runslist, columns = columns)
    #print(runs)
    for t in testfiles:
        path = row[8] + '\\' + t
        try:
            df = pd.read_csv(path,delimiter = '\t')
            df['folder'] = row[8]
            print(path)
            print(df)
            #if len(df.index)== len(codelist):
            runs = pd.merge(runs, df, on=['folder', 'code'])
            #print(runs)
        except FileNotFoundError as e:
            print(e)
        except:
            pass
    dflist.append(runs)
result = pd.concat(dflist)
print(result)
result = result[allcols]
result.to_csv('reputation_results_conserv_new_unrestricted_mem.csv')
result
```
